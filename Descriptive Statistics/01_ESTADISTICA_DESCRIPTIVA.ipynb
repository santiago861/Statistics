```python
# import libraries
library(readxl) # read excel data
library(dplyr)  # data manipulation
```

```python
# read data Base
BASE <- read_excel("BASES/BASE.xlsx")
```

```python
head(BASE)
```

**3. Variable Numerica: Superficie**

```python
print(paste("El maximo de la variable es : ", max(BASE$Superficie)))
print(paste("El minimo de la variable es : ", min(BASE$Superficie)))
print(paste("El promedio de la variable es : ",mean(BASE$Superficie)))
print(paste("La mediana de la variable es : ",median(BASE$Superficie)))
print(paste("El percentile 10% de la variable es : ",quantile(BASE$Superficie, 0.1)))
print(paste("La varianza de la variable es : ",var(BASE$Superficie)))
print(paste("La desviacion estandar de la variable es : ",sd(BASE$Superficie)))
```

**BOXPLOT**

```python
options(repr.plot.width = 5, repr.plot.height = 5)
boxplot(BASE$Superficie, main = "Box plot de la variable Superficie")
```

```python

boxplot(BASE$Superficie, main = "Box plot de la variable Superficie", ylim = c(0,300))
abline( h = quantile(BASE$Superficie,0.25 ), col = "black")
abline( h = quantile(BASE$Superficie,0.75 ), col = "red" )
abline( h = quantile(BASE$Superficie,0.50 ), col = "green" )
RIC = quantile(BASE$Superficie,0.75 ) - quantile(BASE$Superficie,0.25 )
print(paste("El rango intercuartilico es : ", RIC))
brazo_superior <- quantile(BASE$Superficie,0.75 ) + 1.5*RIC
brazo_inferior <- quantile(BASE$Superficie,0.25 ) - 1.5*RIC
abline( h = brazo_superior, col = "purple" )
abline( h = brazo_inferior, col = "blue" )
```
